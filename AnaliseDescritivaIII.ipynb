#### 1. Baixando dados de ativos
    1. Baixe os dados de três ativos da sua escolha do Yahoo Finance, correspondente a um período de pelo menos 2 anos até a data mais recente.
    2. Salve esses dados para que você possa utilizá-los na próxima tarefa.

```python
import yfinance as yf
import pandas as pd
import plotly.express as px

ativos = ["VALE3.SA", "ITUB3.SA", "PETR4.SA"]

dt_ini = "2023-01-01"
dt_fin = "2024-12-31"

dados = yf.download(ativos, start=dt_ini, end=dt_fin)

dados
```

```python
# Salvar os dados em um arquivo CSV
dados.to_csv("dados_acoes.csv")
```

#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
# Calculando a amplitude de variação dentro do dia para cada ativo
for ativo in ativos:
    dados[('Amplitude_Variacao', ativo)] = dados[('High', ativo)] - dados[('Low', ativo)]

# Removendo a hierarquia de colunas para facilitar a visualização
dados.columns = ['_'.join(col).strip() for col in dados.columns.values]

print(dados[[f'High_{ativo}' for ativo in ativos] + [f'Low_{ativo}' for ativo in ativos] + [f'Amplitude_Variacao_{ativo}' for ativo in ativos]])
```

```python
# Resetando o índice para ter a coluna 'Date' disponível
dados = dados.reset_index()

# Criando DataFrame para plotar
df_plot = pd.DataFrame()
for ativo in ativos:
    temp_df = dados[['Date', f'Amplitude_Variacao_{ativo}']].copy()
    temp_df['Ativo'] = ativo
    temp_df = temp_df.rename(columns={f'Amplitude_Variacao_{ativo}': 'Amplitude_Variacao'})
    df_plot = pd.concat([df_plot, temp_df])

# Criando o gráfico interativo com plotly
fig = px.line(df_plot, x='Date', y='Amplitude_Variacao', color='Ativo', labels={'Amplitude_Variacao': 'Amplitude de Variação', 'Date': 'Data'}, title='Variação da Amplitude Diária dos Ativos')

# Habilitar funcionalidades interativas como zoom e pan
fig.update_layout(
    xaxis=dict(
        rangeslider=dict(
            visible=True
        ),
        type='date'
    )
)

# Mostrando o gráfico interativo
fig.show()

```

O ativo que mostrou uma maior volatilidade foi Vale 3. Apresentando um pico em fevereiro de 2023 e maior baixa em abril de 2024.
