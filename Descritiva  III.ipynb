#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
# Ler o arquivo CSV, especificando que os rótulos das colunas estão na segunda linha (header=1)
df = pd.read_csv('dados_acoes.csv', header=1)

# Remover a primeira linha que contém valores de cabeçalho adicionais
df = df.iloc[1:]

# Renomear as colunas para facilitar o acesso
df.columns = ['Date', 'ITUB3_SA_Close', 'PETR4_SA_Close', 'VALE3_SA_Close',
              'ITUB3_SA_High', 'PETR4_SA_High', 'VALE3_SA_High',
              'ITUB3_SA_Low', 'PETR4_SA_Low', 'VALE3_SA_Low',
              'ITUB3_SA_Open', 'PETR4_SA_Open', 'VALE3_SA_Open',
              'ITUB3_SA_Volume', 'PETR4_SA_Volume', 'VALE3_SA_Volume']

# Converter a coluna 'Date' para o tipo datetime
df['Date'] = pd.to_datetime(df['Date'])

# Calcular a média móvel de 90 dias para cada ativo
df['ITUB3_SA_90d_MA'] = df['ITUB3_SA_Close'].rolling(window=90).mean()
df['PETR4_SA_90d_MA'] = df['PETR4_SA_Close'].rolling(window=90).mean()
df['VALE3_SA_90d_MA'] = df['VALE3_SA_Close'].rolling(window=90).mean()

# Exibir as primeiras linhas do DataFrame com as médias móveis
df.head()

````

````python
# Calcular o desvio padrão de 90 dias para cada ativo
df['ITUB3_SA_90d_STD'] = df['ITUB3_SA_Close'].rolling(window=90).std()
df['PETR4_SA_90d_STD'] = df['PETR4_SA_Close'].rolling(window=90).std()
df['VALE3_SA_90d_STD'] = df['VALE3_SA_Close'].rolling(window=90).std()

# Exibir as primeiras linhas do DataFrame com os desvios padrões móveis
df.head()
````

````python
# Criar um gráfico interativo com a média móvel de 90 dias dos valores de fechamento
fig = px.line(df, x='Date', y=['ITUB3_SA_90d_MA', 'PETR4_SA_90d_MA', 'VALE3_SA_90d_MA'],
              labels={'value': 'Média Móvel 90 dias', 'variable': 'Ativo'},
              title='Média Móvel de 90 Dias dos Valores de Fechamento Ajustados')

# Exibir o gráfico
fig.show()
````

````python
# Criar um gráfico interativo com o desvio padrão de 90 dias dos valores de fechamento
fig = px.line(df, x='Date', y=['ITUB3_SA_90d_STD', 'PETR4_SA_90d_STD', 'VALE3_SA_90d_STD'],
              labels={'value': 'Desvio Padrão 90 dias', 'variable': 'Ativo'},
              title='Desvio Padrão em Janela Móvel de 90 Dias dos Valores de Fechamento Ajustados')

# Exibir o gráfico
fig.show()
````

Eu vi como os gráficos mostram as mudanças nos dados financeiros, com altos e baixos. Ao olhar para os gráficos de médias móveis de 90 dias, desvios padrão de 90 dias e a variação diária dos preços de ITUB3.SA, PETR4.SA e VALE3.SA, notei alguns padrões interessantes.

Os períodos de maior volatilidade são claros nos gráficos de desvio padrão e de variação diária. Essas mudanças coincidem, mostrando uma relação direta entre momentos de alta volatilidade e maiores variações diárias.

As médias móveis ajudam a suavizar essas flutuações e mostram tendências que complementam a análise. No geral, os gráficos estão alinhados, indicando que as fases de maior volatilidade aparecem em todas as métricas.
